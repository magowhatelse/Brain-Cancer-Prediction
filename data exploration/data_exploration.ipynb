```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns


df = pd.read_csv(r"C:\Brain Cancer Prediction\data\CSV\data.csv", sep=",")
```

```python

lengths = [len(df[df["Cancer"] == value]) for value in df["Cancer"].unique()]
grades = [i for i in range(3)]
plot = sns.barplot(x=grades,y=lengths).set_title("Distribution of Cancer Grades")
plt.xlabel("Cancer Grades")
plt.ylabel("Count")
plt.show()

#-------------- Train Test split ---------------#
X = df.drop(columns=["Cancer"], inplace=False)
y = df["Cancer"]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)


#-------------- Train Val split ---------------#
X_train, X_valdiation, y_train, y_validation = train_test_split(X_train, y_train, test_size=0.2) 

# plots of the distribution
y_train.value_counts().sort_index().plot(kind="bar")
plt.savefig("train_distribution")
y_test.value_counts().sort_index().plot(kind="bar")
plt.savefig("test_distribution")
y_validation.value_counts().sort_index().plot(kind="bar")
plt.savefig("validation_distribution")
```
